Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
from ipyleaflet import Map, basemaps

# Mapa centrado en Madrid
map = Map(center = (40.35, -3.8), zoom = 11.5, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
map
```

```python
from ipyleaflet import CircleMarker, LayerGroup

# 1. Limpieza: Reemplazamos comas por puntos y convertimos a float
df_mapa = df_cinturon_sur.dropna(subset=['latitude', 'longitude']).copy()

# Función para limpiar las coordenadas si vienen con coma
def limpiar_coordenada(valor):
    if isinstance(valor, str):
        return float(valor.replace(',', '.'))
    return float(valor)

df_mapa['latitude'] = df_mapa['latitude'].apply(limpiar_coordenada)
df_mapa['longitude'] = df_mapa['longitude'].apply(limpiar_coordenada)

# 2. Configuración de colores
poblaciones = df_mapa['level5'].unique()
colores_lista = ['red', 'blue', 'green', 'orange', 'purple']
colores_dict = dict(zip(poblaciones, colores_lista))

# 3. Crear la capa y añadir los puntos
capa_puntos = LayerGroup()

for _, fila in df_mapa.iterrows():
    pob = fila['level5']
    color_pob = colores_dict.get(pob, 'black')
    
    marcador = CircleMarker(
        location=(fila['latitude'], fila['longitude']),
        radius=3,
        color=color_pob,
        fill_color=color_pob,
        fill_opacity=0.7,
        weight=1
    )
    capa_puntos.add_layer(marcador)

# 4. Ajustar el mapa y añadir capa
map.center = (df_mapa['latitude'].mean(), df_mapa['longitude'].mean())
map.zoom = 11

# Limpiar capas previas si existen
if len(map.layers) > 1:
    map.layers = (map.layers[0],) 

map.add_layer(capa_puntos)
map
```
